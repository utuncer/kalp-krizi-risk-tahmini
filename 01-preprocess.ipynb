## Verileri Okuma

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
df = pd.read_csv(r"data\heart.csv")
df.head(10)
```

```python
df.isnull().sum()
```

 ## Verilerin Sayısal Değerlere Dönüştürülmesi 
 - Verilerin sayısal değerlere dönüştürülmesi işlemi (Categorical Encoding) yapılıyor

```python
categorical_encoding = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]

for column in categorical_encoding:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    
    print(f"--- {column} ---")
    for label, value in zip(le.classes_, range(len(le.classes_))):
        print(f"'{label}' : {value}")
    print("-" * 15)

df.head(10)
```

## Verilere Standardizasyon İşlemi

- Model büyük sayılı verilerin kalp krizi riski için daha önemli olduğunu sanabilir  
- Sayısal verileri ortalaması 0, standart sapması 1 olacak şekilde güncelliyoruz

```python
numeric_columns = ["Age","RestingBP","Cholesterol","MaxHR","Oldpeak"]

scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
df.head(10)
```

## Model Eğitimi için Hazırlanmış Verilerin Çıktısını Alıyorum

```python
df.to_csv(r"data\heart_processed.csv", index= False)
```
